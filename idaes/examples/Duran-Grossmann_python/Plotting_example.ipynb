Data from Systematic Methods of Chemical Process Design, Biegler, Grossmann, Westerberg
page 355

```python
from idaes.functions.DuranGrossmann import generate_curvesKW, CurveData
```

```python
CD = CurveData(0)
CD.Cooling_Tin = [160,249]
CD.Cooling_Tout = [93,138]
CD.Cooling_Q = [-588.9,-1171.1]
CD.Heating_Tin = [60,116]
CD.Heating_Tout = [160,260]
CD.Heating_Q = [762.0,875.5]

generate_curvesKW(CD)
```
